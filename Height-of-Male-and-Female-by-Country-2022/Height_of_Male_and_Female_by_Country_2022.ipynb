```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data info
data = pd.read_csv('Height of Male and Female by Country 2022.csv', index_col='Country Name', encoding='latin-1')
data.info()
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(data).describe()
```

```python
#Fifty tallest countries in 2022 (Males)
plt.figure(figsize=(25, 15))
male = data['Male Height in Cm'].head(50).plot(kind='bar', color = 'green')
plt.title('Fifty tallest countries in 2022 (Males)', fontsize=25)
plt.xlabel('Countries, -', fontsize=18)
plt.ylabel('Average High, Cm', fontsize=18)
plt.ylim(ymin = 175, ymax = 185)
plt.show(male)
```

```python
#Fifty tallest countries in 2022 (Females)
plt.figure(figsize=(25, 15))
data_sorted = data.sort_values('Female Height in Cm', ascending=False)
female = data_sorted['Female Height in Cm'].head(50).plot(kind='bar', color = 'pink')
plt.title('Fifty tallest countries in 2022 (Females)', fontsize=25)
plt.xlabel('Countries, -', fontsize=18)
plt.ylabel('Average High, Cm', fontsize=18)
plt.ylim(ymin = 162, ymax = 171)
plt.show(female)
```

```python
#Fifty shortest countries in 2022 (Males)
plt.figure(figsize=(25, 15))
shortest_males = data.sort_values('Male Height in Cm', ascending=True)
male = shortest_males['Male Height in Cm'].head(50).plot(kind='bar', color = 'blue')
plt.title('Fifty tallest countries in 2022 (Males)', fontsize=25)
plt.xlabel('Countries, -', fontsize=18)
plt.ylabel('Average High, Cm', fontsize=18)
plt.ylim(ymin = 158, ymax = 170)
plt.show(male)
```

```python
#Fifty shortest countries in 2022 (Females)
plt.figure(figsize=(25, 15))
shortest_females = data.sort_values('Female Height in Cm', ascending=True)
female = shortest_females['Female Height in Cm'].head(50).plot(kind='bar', color = 'red')
plt.title('Fifty shortest countries in 2022 (Females)', fontsize=25)
plt.xlabel('Countries, -', fontsize=18)
plt.ylabel('Average High, Cm', fontsize=18)
plt.ylim(ymin = 150, ymax = 159)
plt.show(female)
```
